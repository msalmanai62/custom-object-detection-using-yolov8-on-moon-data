```python
import yaml
from ultralytics import YOLO

# Load the configuration file
with open('data.yaml', 'r') as file:
    config = yaml.safe_load(file)

hyperparameters = {
#     'lr0': 0.00269,
# 'lrf': 0.00288,
# 'momentum': 0.73375,
# 'weight_decay': 0.00015,
# 'warmup_epochs': 1.22935,
# 'warmup_momentum': 0.1525,
# 'box': 18.27875,
# 'cls': 1.32899,
# 'dfl': 0.56016,
# 'hsv_h': 0.01148,
# 'hsv_s': 0.53554,
# 'hsv_v': 0.13636,
# 'degrees': 0.0,
# 'translate': 0.12431,
'scale': 0.07643,
'shear': 0.0,
'perspective': 0.0,
'flipud': 0.0,
'fliplr': 0.08631,
'mosaic': 0.42551,

}
# Initialize the model (YOLOv8)
model = YOLO('yolov8m.pt')  # You can use a different YOLO model (yolov8s.pt, yolov8m.pt, etc.)
data_file_path = "/content/drive/MyDrive/moon detection/data.yaml"
model.train(data=data_file_path, epochs=50, **hyperparameters)
```

```python
model.save('trained_moon_model.pt')
```

```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt

# Load the trained model
# model = YOLO('yolov8_trained_moon.pt')


# IMG-20240624-WA0008.jpg
# Load an image
image_path = "/content/drive/MyDrive/moon detection/monn_data/data/images/IMG-20240624-WA0004_jpg.rf.96838c3efa7c8c8e9d26f927cd1e8c11.jpg"
image = cv2.imread(image_path)

# Perform inference
results = model.predict(source=image, save=True)
plt.figure(figsize=(10, 10))
# Plot the image with bounding boxes
plt.imshow(cv2.cvtColor(results[0].plot(), cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
import os

from ultralytics import YOLO
import cv2


VIDEOS_DIR = os.path.join('.', 'videos')
video_path = "/content/drive/MyDrive/moon detection/moon_test_2.mp4"

# video_path = os.path.join(VIDEOS_DIR, 'alpaca1.mp4')
video_path_out = '{}_out.mp4'.format(video_path)

cap = cv2.VideoCapture(video_path)
ret, frame = cap.read()
H, W, _ = frame.shape
out = cv2.VideoWriter(video_path_out, cv2.VideoWriter_fourcc(*'MP4V'), int(cap.get(cv2.CAP_PROP_FPS)), (W, H))

model_path = os.path.join('.', 'runs', 'detect', 'train', 'weights', 'last.pt')

# Load a model
# model = YOLO("yolov8_trained.pt")  # load a custom model

threshold = 0.4

while ret:

    results = model(frame)[0]

    for result in results.boxes.data.tolist():
        x1, y1, x2, y2, score, class_id = result

        if score > threshold:
            cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 4)
            cv2.putText(frame, results.names[int(class_id)].upper(), (int(x1), int(y1 - 10)),
                        cv2.FONT_HERSHEY_SIMPLEX, 1.3, (0, 255, 0), 3, cv2.LINE_AA)

    out.write(frame)
    ret, frame = cap.read()

cap.release()
out.release()
cv2.destroyAllWindows()
```
